```python
import pandas as pd
```

```python
MODELS = ["gemini", "claude", "gpt", "gemma-12b", "olmo-7b", "gemma-4b"]
REJECTION_STRATEGIES = ["aggressive", "majority"]
```

```python
for model in MODELS:
    for rejection_strategy in REJECTION_STRATEGIES:
        df = pd.read_csv(f"../data/collab-rec-2026/analysis/{model}_mami_{rejection_strategy}_scores.csv")
        # df = df.loc[df["round_nr"] <=4] # early stopping after 4 rounds
        avg_time = df["time_taken"].mean()
        print(f"Model: {model}, Rejection Strategy: {rejection_strategy}, Average Time Taken: {avg_time:.2f} seconds")
```

```python
for model in MODELS:
    for rejection_strategy in REJECTION_STRATEGIES:
        df = pd.read_csv(f"../data/collab-rec-2026/analysis/{model}_mami_{rejection_strategy}_scores.csv")
        df = df.loc[(df["agent_name"]=="moderator") & (df["round_nr"]==10)]
        avg_time = df["total_token_count"].mean()
        print(f"Model: {model}, Rejection Strategy: {rejection_strategy}, Average token count: {avg_time:.2f}")
```

```python
for model in MODELS:
    for rejection_strategy in REJECTION_STRATEGIES:
        df = pd.read_csv(f"../data/collab-rec-2026/analysis/{model}_mami_{rejection_strategy}_scores.csv")
        # df = df.loc[df["round_nr"] <=4] # early stopping after 4 rounds
        avg_hall = df.groupby('round_nr')['hallucination_rate'].mean()
        print(f"Model: {model}, Rejection Strategy: {rejection_strategy}, Average Hallucination: {avg_hall}")
```
