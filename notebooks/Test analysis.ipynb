```python
import json
import pandas as pd
```

```python
def load_queries(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)
```

```python
file_name = "../data/collab-rec-2026/llm-results/olmo-7b/sasi/olmo-7b_sasi.json"
responses = load_queries(file_name)
len(responses)
```

```python
filtered = [
    q for q in responses
    if q.get("response")
]
print(len(filtered))
```

```python
filtered = [
    q for q in responses
    if q.get("response")
]
print(len(filtered))
with open(file_name, "w", encoding="utf-8") as f:
    json.dump(filtered, f, indent=2, ensure_ascii=False)
print("saved filtered responses to", file_name)
```
