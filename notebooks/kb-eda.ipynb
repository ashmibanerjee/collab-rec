```python
def filter_kb(kb, config):
    """
    Sequentially filters the KB DataFrame based on available config filters.
    Matching is case-insensitive.
    Missing filters are ignored.
    """

    df = kb.copy()
    filters = config.get("filters", {})

    for key, value in filters.items():
        if value is None:
            continue

        if key not in df.columns:
            continue

        df = df[
            df[key]
            .astype(str)
            .str.lower()
            .eq(value.strip().lower())
        ]

    return df.city.unique().tolist()
filter_kb(kb, sample_config)
```

```python
for col in ["low_season", "medium_season", "high_season"]:
    kb[col] = kb[col].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) else x
    )


MONTHS = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
def expand_seasonality_row(row):
    month_map = {month: None for month in MONTHS}

    for month in row["low_season"]:
        month_map[month] = "low"

    for month in row["medium_season"]:
        month_map[month] = "medium"

    for month in row["high_season"]:
        month_map[month] = "high"

    return month_map
```

```python
seasonality_expanded = kb.apply(expand_seasonality_row, axis=1, result_type="expand")

kb_expanded = pd.concat(
    [kb[["city"]], seasonality_expanded],
    axis=1
)
kb_expanded
```
