```python
import pandas as pd
```

```python
MODEL_NAME = "claude"
REJECTION_STRATEGY = "aggressive"
```

```python
df = pd.read_csv(f"../data/collab-rec-2026/analysis/{MODEL_NAME}_mami_{REJECTION_STRATEGY}_10_rounds_scores.csv")
df.head()
```

```python
grouped_df = df.groupby("round_nr").agg({"success_score": "mean"}).reset_index()
grouped_df.head()
```

```python
grouped_df["delta"] = grouped_df["success_score"].diff()
grouped_df
```

```python
grouped_df.delta.describe()
```
