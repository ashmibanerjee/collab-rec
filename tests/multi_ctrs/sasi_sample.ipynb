```python
import pandas as pd 
import numpy as np 
import os 
import sys
import ast 
```

```python
df = pd.read_csv("../../data/conv-trs/multi-agent/results/sasi_results.csv")
```

```python
df.head()
```

```python
MODELS = ['Gemini2Flash', 'Claude3Point7Sonnet', 'DeepSeekReasoner']
```

```python
common_cities = [] 

for i, row in df.iterrows(): 
    cities = set(ast.literal_eval(row[f'response_{MODELS[0]}'])) & set(ast.literal_eval(row[f'response_{MODELS[1]}'])) & set(ast.literal_eval(row[f'response_{MODELS[2]}']))
    common_cities.append(cities)

df['common_cities'] = common_cities
    
```

```python
df.head()
```

```python
def len_list(common_cities):
    return len(common_cities)

print(np.mean(df['common_cities'].apply(len_list)))
```

```python
def find_level_pop(config_id):
    pop_level = config_id.split("_")[4:]
    return "_".join(pop_level)

df['pop_level'] = df['config_id'].apply(find_level_pop)
```

```python
df['len_common_cities'] = df['common_cities'].apply(len)
df1 = df.groupby('pop_level')['len_common_cities'].mean()
```
